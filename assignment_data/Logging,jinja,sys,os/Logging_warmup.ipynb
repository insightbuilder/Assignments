```python
import logging

logging.basicConfig(filename='log5.log', level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')
                    # date_format=,
                    # level)

for i in range(0,15):
    if(i%2 ==0):
        logging.warning('Log warning message- multiple of 2')
    elif(i%3==0):
        print(i)  
        logging.info(f"Number is {i}")      
        logging.critical('Log critical message- multiple of 3')
    else:
        logging.error('Log error message - Elsee')
        

# Logging levels ----> Debug(10) < Info < Warning(30) < Error < Critical(50)
```
